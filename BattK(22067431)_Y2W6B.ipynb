```python
#3
def func(x):
    return sin(x)*tan(x)

Npoint = 100
fi_sum = func((np.pi/4)*np.random.random(Npoint))
o =np.sum(fi_sum)

b = np.pi/4
a = 0
N = Npoint
result = (b - a) / N * o
print(result)

#Exact answer via wolframalpha is 1.743

x = result
y = N

plt.plot(x,y)
plt.xlabel('Monte Carlo Value at point N')
plt.ylabel('N points')
plt.title('Graph of MC value was N increases')
plt.show()

'''
I cant figure out how to increase the number of N points and find
a resulting MC answer verses that N point, if i could i would make the
MC the y axis and resulting MC the x axis.
'''
```
